```python
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
```

```python
wondu_mat = count_vect.fit_transform(wondu['아로마'])
print(wondu_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
wondu_sim = cosine_similarity(wondu_mat, wondu_mat)
print(wondu_sim.shape)
print(wondu_sim[:2])
```

```python
# [:, ::-1] axis = 1 기준으로 2차원 numpy 배열 뒤집음
#코사인 유사도가 측정된 genre_sim 객체를 Numpy의 argsort() 함수를 통해 유사도가 높은 순 정렬
#또한, 각 레코드에서 맨 뒤에 있는 인덱스 값을 비교 행 위치로 뒤집어 
#결과 값을 보자면 첫 번째(0번) 레코드 의 경우 자기 자신 0번을 제외하면 3494번 레코드가 가장 유사도가 높고, 
#그 다음이 813번 레코드, 가장 유사도가 낮은 레코드는 2401번 레코드라는 뜻
wondu_sim_sorted_ind = wondu_sim.argsort()[:, ::-1]
print(wondu_sim_sorted_ind[:1])
```

```python
def find_sim_movie(df, sorted_ind, wondu_id, top_n=10):
    
    # 인자로 입력된 movies_df DataFrame에서 'wondu' 컬럼이 입력된 title_name 값인 DataFrame추출
    wondu_good = df[df['원두이름'] == wondu_id]
    
    # title_named을 가진 DataFrame의 index 객체를 ndarray로 반환하고 
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n 개의 index 추출
   
    wondu_index = wondu_good.index.values
    similar_indexes = sorted_ind[wondu_index, :(top_n)]
    
    # 추출된 top_n index들 출력. top_n index는 2차원 데이터 임. 
    #dataframe에서 index로 사용하기 위해서 1차원 array로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

```python
similar_wondu = find_sim_movie(wondu, wondu_sim_sorted_ind, '에티오피아',10)
similar_wondu[['원두이름', '산도']]
```

```python
percentile = 0.6
m = wondu['랭킹'].quantile(percentile)
C = wondu['산도'].mean()

def weighted_vote_average(record):
    v = record['랭킹']
    R = record['산도']
    
    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )   

wondu['굿랭킹'] = wondu.apply(weighted_vote_average, axis=1)
```

```python
wondu[['원두이름','산도','굿랭킹','랭킹']].sort_values('굿랭킹',ascending=False)[:10]
                                                                          
```

```python
def find_sim_wondu(df, sorted_ind, wondu_name, top_n=10):
    wondu_title = df[df['원두이름'] == wondu_name]
    wondu_index = wondu_title.index.values
    
    # top_n의 2배에 해당하는 쟝르 유사성이 높은 index 추출 
    similar_indexes = sorted_ind[wondu_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
# 기준 영화 index는 제외
    similar_indexes = similar_indexes[similar_indexes != wondu_index]
    
    # top_n의 2배에 해당하는 후보군에서 weighted_vote 높은 순으로 top_n 만큼 추출 
    return df.iloc[similar_indexes].sort_values('굿랭킹', ascending=False)[:top_n]

similar_wondu = find_sim_wondu(wondu, wondu_sim_sorted_ind, 2,10)
similar_wondu[['원두이름', '산도', '굿랭킹']]
```
